```python
#checking different levels of door numbers and cylindernumber
cars.doornumber.astype('category').value_counts(),cars.cylindernumber.astype('category').value_counts()
```

```python
#Function to map the cotegorical levels to actual numbers
def num_map(x):
    return x.map({'four':4,'six':6,'five':5,'eight':8,'two':2,'twelve':12,'three':3})
```

```python
#applying function
cars[["doornumber","cylindernumber"]] = cars[["doornumber","cylindernumber"]].apply(num_map)
```

```python
#applying function to cars_original
cars_original[["doornumber","cylindernumber"]] = cars_original[["doornumber","cylindernumber"]].apply(num_map)
```

```python
cars_categorical = cars_original.select_dtypes(include=['object'])
```

```python
cars_categorical.head()
```

```python
cars[["doornumber",'cylindernumber']].head()
```

```python
#splitting company name from carname in car_categorical
def split_company(x):
    k=[]
    for i in range(len(x)):
        y = x.values[i].split(" ")
        k.append(y[0])
    return k
CName = pd.Series(split_company(cars_categorical.CarName))
CName.value_counts().sum()
```

```python
#better function to do above
def split_company2(x):
    return x.split(" ")[0]

CompanyName = cars_categorical.CarName.apply(split_company2)# apply() function applies the function to earch series element
CompanyName.value_counts()
```

```python
#insert CarName into cars_categorical
cars_categorical.insert(0,"CompanyName",CompanyName)
cars_categorical.head()
```

```python
#Drop CarName
cars_categorical.drop(['CarName'],axis=1,inplace=True)
```

```python
cars_categorical.head()
```

```python
cars_categorical.CompanyName.value_counts()
```

```python
cars_categorical.CompanyName.unique()
```

```python
#Replace wrong CompanyName with correct one
#1.convert all names to lowercase
cars_categorical.CompanyName = cars_categorical.CompanyName.str.lower()
```

```python
#Replace function
def replace_name(a,b):
    cars_categorical.CompanyName.replace(a,b,inplace=True)
```

```python
replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')
```

```python
cars_categorical.CompanyName.value_counts().sum()
```

```python
#convert into dummies
cars_dummies=pd.get_dummies(cars_categorical,drop_first = True)
```

```python
cars_dummies.head(),cars_dummies.shape
```

```python
#drop categorical_columns from cars
cars = cars.drop(list(cars_original.select_dtypes(include=['object']).columns),axis=1)
```

```python
#concatinate dummies with new cars
cars = pd.concat([cars,cars_dummies],axis=1)
```

# 3. Model building and evaluation

```python
#Split train, test data
from sklearn.model_selection import train_test_split
df_train,df_test = train_test_split(cars,train_size=0.7,test_size=0.3,random_state=100)
```

```python
cars_numeric = cars_original.select_dtypes(exclude=['object'])
```

```python
#All numeric columns in cars+ cylindernumber and doornumber
cars_numeric.columns
```

```python
#importing standardscalar()
from sklearn.preprocessing import StandardScaler

#creating a scaling object
scaler = StandardScaler()

#creating a list of variables you need to scale
varlist = ['symboling', 'doornumber', 'wheelbase', 'carlength',
       'carwidth', 'carheight', 'curbweight', 'cylindernumber', 'enginesize',
       'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm',
       'citympg', 'highwaympg', 'price']

#scaling these variables using 'fit_transform'
df_train[varlist]=scaler.fit_transform(df_train[varlist])
```

```python
df_train.head()
```

```python
#Split train into x and y
y_train=df_train.pop('price')
x_train= df_train
```

## Building the first model with all the variables 

```python
from sklearn.linear_model import LinearRegression
#instantiate
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
#print(lm.coef)
#print(lm.intercept)
```

## Model building using RFE

```python
#import RFE
from sklearn.feature_selection import RFE

#RFE with 15 features
rfe1 = RFE(lm,15)

#fit with 15 features
rfe1.fit(x_train, y_train)

#print the boolean result
print(rfe1.support_)
```

```python
print(rfe1.ranking_)
```

## Model building and evaluation

```python
#import statsmodels
import statsmodels.api as sm

#subset the features selected by rfe1
col1=x_train.columns[rfe1.support_]

#subsetting training data from 15 selected columns
x_train_rfe1=x_train[col1]

#add a constant to the model
x_train_rfe1 = sm.add_constant(x_train_rfe1)
x_train_rfe1.head()
```

```python
#Fitting the model with 15 variables
lm1 = sm.OLS(y_train,x_train_rfe1).fit()
print(lm1.summary())
```

### Check VIF values of feature variables

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif=pd.DataFrame()
vif['Features'] = x_train_rfe1.columns
vif['VIF']=[variance_inflation_factor(x_train_rfe1.values,i) for i in range(x_train_rfe1.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```
